```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

# import streamlit as st
# import altair as alt
# import plotly.express as px
```

```python
monsters = pd.read_json('../data/creatures/srd_5e_monsters.json')
```

```python
monsters.columns
```

```python
monsters_num = monsters[['name', 'Hit Points', 'Speed', 'STR', 'STR_mod',
       'DEX', 'DEX_mod', 'CON', 'CON_mod', 'INT', 'INT_mod', 'WIS', 'WIS_mod',
       'CHA', 'CHA_mod']]
```

```python
for i in monsters.columns:
    # print(i)
    data = monsters[i].value_counts().to_frame().reset_index()

    fig, ax = plt.subplots(figsize=(20, 5))
    sns.barplot(data=data.head(10), x='index', y=i)
    ax.tick_params(axis='x', rotation=45)
```

```python
monsters.plot(subplots=True, layout=(4,5), figsize=(20, 20))
plt.tight_layout()
plt.show()
```
